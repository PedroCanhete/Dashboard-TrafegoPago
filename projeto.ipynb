```python
import pandas as pd
dados = pd.read_csv("Material/website_wata.csv")
dados.head()
```

```python
dados.describe()
```

```python
dados.info()
```

```python
# Criando os minutos e segundos para Session Duration
dados["Minutos_Session"] = dados["Session Duration"].astype(int)
dados["Segundos_Session"] = ((dados["Session Duration"] - dados["Minutos_Session"]) * 60) / 100
dados["Tempo Float Session"] = dados["Minutos_Session"] + dados["Segundos_Session"]

# Criando os minutos e segundos para Time on Page
dados["Minutos_Time"] = dados["Time on Page"].astype(int)
dados["Segundos_Time"] = ((dados["Time on Page"] - dados["Minutos_Time"]) * 60) / 100
dados["Tempo Float Time"] = dados["Minutos_Time"] + dados["Segundos_Time"]

# Removendo colunas temporárias e originais
dados = dados.drop(columns=["Session Duration", "Time on Page", 
                            "Minutos_Session", "Segundos_Session", 
                            "Minutos_Time", "Segundos_Time"])

# Renomeando colunas finais
dados.rename(columns={'Tempo Float Session': 'Duração da sessão (minutos)',
                      'Tempo Float Time': 'Tempo na página (minutos)', 
                      'Page Views': 'Páginas visitadas',
                      'Bounce Rate': 'Taxa de rejeição', 
                      'Traffic Source': 'Origem do tráfego',
                      'Previous Visits': 'Visitas anteriores',
                      'Conversion Rate': 'Taxa de conversão'}, inplace=True)

dados.head()
```
